```python
import numpy as np
import matplotlib.pyplot as plt
import pyabf
import os
import glob
import scipy.signal
import pandas as pd
path = "C:\\Users\\SMest\\Downloads\\VC_K channel characterization\\"

abf_files = glob.glob(path + "*.abf")
abf = pyabf.ABF(abf_files[0])
```

```python
abf.setSweep(abf.sweepList[-1])
diff = np.nonzero(np.diff(abf.sweepC))[0]


wave1_bounds = (diff[1], diff[2])
wave2_bounds = (diff[-2], diff[-1])
OFFSET = -70
print(wave1_bounds, wave2_bounds)
```

```python
def subtract_waveforms(x, y, c):
    wave1 = y[wave1_bounds[0]:wave1_bounds[1]]
    wave2 = y[wave2_bounds[0]:wave2_bounds[1]]
    wave_diff = wave1 - wave2
    return x[wave1_bounds[0]:wave1_bounds[1]], wave_diff, c[wave1_bounds[0]:wave1_bounds[1]] + OFFSET
   
```

```python
abf_files_data_out = []
abf_file_id = []
for file in abf_files:
    abf = pyabf.ABF(file)
    plt.clf()
    plt.figure(figsize=(15, 10))
    temp_data = []
    amp_C = []
    for sweep in abf.sweepList:
        abf.setSweep(sweep)
        filter = scipy.signal.butter(3, Wn=100, output='sos', btype='lowpass', analog=False, fs=abf.dataRate)
        x, y, c = subtract_waveforms(abf.sweepX, scipy.signal.sosfiltfilt(filter, abf.sweepY), abf.sweepC)
        plt.subplot(1,3,1)
        plt.plot(x, y)
        plt.xlim(0.23,0.24)
        plt.subplot(1,3,2)
        plt.plot(x, c)
        plt.subplot(1,3,3)
        plt.plot(x, y - c)
        temp_data.append(np.nanmax(y))
        amp_C.append(np.amax(c))
    plt.show()
    abf_files_data_out.append(temp_data)
    abf_file_id.append(file)
    



df = pd.DataFrame(np.vstack(abf_files_data_out).T, columns=abf_file_id, index=amp_C)
df.to_csv(path + "VC_K_channel_characterization.csv")
```
